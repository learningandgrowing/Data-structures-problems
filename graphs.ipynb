```python
#kruskal
class Edge:
    def __init__(self, src, dest, wt):
        self.src = src
        self.dest = dest
        self.wt = wt
def getparent(v, parent):
    if v == parent[v]:
        return v
    return getparent(parent[v], parent)
def kruskal(edges, V):
        
    parent = [i for i in range(V)]
    edges = sorted(edges, key= lambda edge:edge.wt)
    output = []
    count = 0
    i = 0
        
    while count < (V - 1):
        curredge = edges[i]
        srcparent = getparent(curredge.src, parent)
        destparent = getparent(curredge.dest, parent)
        if srcparent != destparent:
            output.append(curredge)
            count += 1
            parent[srcparent] = destparent
        i += 1
    return output
            
        
V, E = [int(i) for i in input().split()]
edges = []
for i in range(E):
    src, dest, wt = [int(i) for i in input().split()]
    edge = Edge(src, dest, wt)
    edges.append(edge)

output = kruskal(edges, V)
for edge in output:
    if edge.src < edge.dest:
        print(edge.src,edge.dest,edge.wt, end='')
        print()
    else:
        print(edge.dest,edge.src,edge.wt, end='')
        print()
    
    
```

```python
#prims
import sys
class Graph:
    def __init__(self,nvertices):
        self.nvertices = nvertices
        
        self.adjmatrices = [[0 for i in range(nvertices)] for j in range(nvertices)]
    def addEdge(self, v1, v2, wt):
        
        
        self.adjmatrices[v1][v2]= wt
        self.adjmatrices[v2][v1] = wt
    def __str__(self):
        return str(self.adjmatrices)
    def __getminverteces(self, weight, visited):
        min_vertex = -1
        for i in range(self.nvertices):
            if (visited[i] is False and (min_vertex == -1 or weight[min_vertex] > weight[i])):
                min_vertex = i
        return min_vertex
    def prims(self):
        visited = [False for i in range(self.nvertices)]
        parent = [-1 for i in range(self.nvertices)]
        weight = [sys.maxsize for i in range(self.nvertices)]
        weight[0] = 0
        for i in range(self.nvertices-1):
            min_vertex = self.__getminverteces(weight, visited)
            visited[min_vertex] = True
            for j in range(self.nvertices):
                if self.adjmatrices[min_vertex][j]>0 and visited[j] is False:
                    if weight[j] > self.adjmatrices[min_vertex][j]:
                        weight[j] = self.adjmatrices[min_vertex][j]
                        parent[j] = min_vertex
        for i in range(1, self.nvertices):
            if i < parent[i]:
                print(str(i) + " " + str(parent[i]) + " " + str(weight[i]))
                
            else:
                print(str(parent[i]) + " " + str(i) + " " + str(weight[i]))
                
                
            
        

        
V, E = [int(i) for i in input().split()]
g = Graph(V)
for i in range(E):
    v1, v2, wt = [int(i) for i in input().split()]
    g.addEdge(v1, v2, wt)
g.prims()
```
