```python
#midpoint of linked list
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def length(head):
    count = 0
    while head is not None:
        head = head.next
        count += 1
    return count
    
def midpoint_linkedlist(head):
    #  Given a linked list, find and return the midpoint.
    mid = (length(head)-1)//2
    i = 0
    tail = head
    while i < mid:
        tail = tail.next
        i += 1
    
    return tail
        

def ll(arr):
    if len(arr)==0:
        return None
    head = Node(arr[0])
    last = head
    for data in arr[1:]:
        last.next = Node(data)
        last = last.next
    return head

# Main
# Read the link list elements including -1
arr=list(int(i) for i in input().strip().split(' '))
# Create a Linked list after removing -1 from list
l = ll(arr[:-1])
node = midpoint_linkedlist(l)
if node:
    print(node.data)
```
